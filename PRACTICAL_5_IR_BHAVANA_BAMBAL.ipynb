Aim: Representation of Text document in Vector Space Model and computing similarity between two document

```python
#BHAVANA BAMBAL(CS23131)
import math
import string

# ---------------------------------
# Step 1: Define 3 Documents
# ---------------------------------

doc1 = "Information retrieval is fun and interesting"
doc2 = "Information processing and retrieval systems"
doc3 = "Retrieval of information from systems is important"

documents = [doc1, doc2, doc3]

# ---------------------------------
# Step 2: Define Stopwords
# ---------------------------------

stopwords = {"is", "and", "of", "from", "the", "a", "an"}

# ---------------------------------
# Step 3: Preprocess Documents
# (Lowercase, remove punctuation,
# tokenization, remove stopwords)
# ---------------------------------

processed_docs = []

for doc in documents:
    doc = doc.lower()  # convert to lowercase
    doc = doc.translate(str.maketrans("", "", string.punctuation))  # remove punctuation
    words = doc.split()  # tokenization
    words = [word for word in words if word not in stopwords]  # remove stopwords
    processed_docs.append(words)

# ---------------------------------
# Step 4: Create Vocabulary
# (Alphabetical Order)
# ---------------------------------

vocab = sorted(set(word for doc in processed_docs for word in doc))

print("Vocabulary (Alphabetical Order):")
print(vocab)

# ---------------------------------
# Step 5: Create Document Vectors
# (1 if word present, else 0)
# ---------------------------------

vectors = []

for doc in processed_docs:
    vector = []
    for word in vocab:
        if word in doc:
            vector.append(1)
        else:
            vector.append(0)
    vectors.append(vector)

# Print Document Vectors
for i in range(len(vectors)):
    print(f"\nDocument {i+1} Vector:")
    print(vectors[i])

# ---------------------------------
# Step 6: Cosine Similarity Function
# ---------------------------------

def cosine_similarity(v1, v2):
    dot_product = sum(a*b for a, b in zip(v1, v2))
    magnitude1 = math.sqrt(sum(a*a for a in v1))
    magnitude2 = math.sqrt(sum(b*b for b in v2))

    if magnitude1 == 0 or magnitude2 == 0:
        return 0

    return dot_product / (magnitude1 * magnitude2)

# ---------------------------------
# Step 7: Calculate Similarities
# ---------------------------------

print("\nCosine Similarities:")

print("D1 & D2:", cosine_similarity(vectors[0], vectors[1]))
print("D1 & D3:", cosine_similarity(vectors[0], vectors[2]))
print("D2 & D3:", cosine_similarity(vectors[1], vectors[2]))
```
