```python
X = df.drop(columns=['target_class'])
y = df['target_class']
```

```python
X.head()
```

```python
X = pd.get_dummies(X,dtype = int)
```

```python
X.head()
```

```python
x =df.iloc[:, :-1]
y =df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.7,test_size=0.3,random_state=42)
X_train.shape,X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
svc=SVC()
svc.fit(X_train,y_train)
y_pred=svc.predict(X_test)
print(f'Model accuracy score with default hyperparameters:\n{accuracy_score(y_test,y_pred):.4f}')
```

```python
svc=SVC(C=100.0)
```

```python
svc.fit(X_train,y_train)
y_pred=svc.predict(X_test)
```

```python
print('Model accuracy score with rbf kernel and C=100.0: {0:.4f}'.format(accuracy_score(y_test,y_pred)))
```

```python
linear_svc=SVC(kernel='linear',C=1.0)
linear_svc.fit(X_train,y_train)
y_pred=linear_svc.predict(X_test)
print('Model accuracy score with linear kernel and C=1.0: {0:.4f}'.format(accuracy_score(y_test,y_pred)))
```

```python
linear_svc=SVC(kernel='linear',C=100.0)
linear_svc.fit(X_train,y_train)
y_pred=linear_svc.predict(X_test)
print('Model accuracy score with linear kernel and C=100.0: {0:.4f}'.format(accuracy_score(y_test,y_pred)))
```

```python
linear_svc=SVC(kernel='linear',C=1000.0)
linear_svc.fit(X_train,y_train)
y_pred=linear_svc.predict(X_test)
print('Model accuracy score with linear kernel and C=1000.0: {0:.4f}'.format(accuracy_score(y_test,y_pred)))
```

```python
poly_svc=SVC(kernel='poly',C=1.0)
poly_svc.fit(X_train,y_train)
y_pred=poly_svc.predict(X_test)
print('Model accuracy score with polynomial kernel and C=1.0: {0:.4f}'.format(accuracy_score(y_test,y_pred)))
```

```python
poly_svc=SVC(kernel='poly',C=100.0)
poly_svc.fit(X_train,y_train)
y_pred=poly_svc.predict(X_test)
print('Model accuracy score with polynomial kernel and C=100.0: {0:.4f}'.format(accuracy_score(y_test,y_pred)))
```

```python
poly_svc=SVC(kernel='poly',C=1000.0)
poly_svc.fit(X_train,y_train)
y_pred=poly_svc.predict(X_test)
print('Model accuracy score with polynomial kernel and C=1000.0: {0:.4f}'.format(accuracy_score(y_test,y_pred)))
```
