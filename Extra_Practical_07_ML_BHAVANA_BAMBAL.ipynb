```python
X = df.drop(columns=['exam_score'])
y = df['exam_score']
```

```python
X.head()
```

```python
X = pd.get_dummies(X,dtype = int)
```

```python
X.head()
```

```python
x =df.iloc[:, :-1]
y =df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.7,test_size=0.3,random_state=42)
X_train.shape,X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.svm import SVR
from sklearn.metrics import r2_score

svr = SVR()
svr.fit(X_train, y_train)
y_pred = svr.predict(X_test)

print(f'Model R2 Score with default hyperparameters:\n{r2_score(y_test, y_pred):.4f}')
```

```python
svr=SVR(C=100.0)
```

```python
from sklearn.svm import SVR
from sklearn.metrics import r2_score

# Instantiate SVR with C=100.0, similar to what was attempted with SVC
svr = SVR(C=100.0)
svr.fit(X_train, y_train)
y_pred = svr.predict(X_test)

print(f'Model R2 Score with C=100.0:\n{r2_score(y_test, y_pred):.4f}')
```

```python
from sklearn.metrics import r2_score
print('Model R2 score with rbf kernel and C=100.0: {0:.4f}'.format(r2_score(y_test,y_pred)))
```

```python
from sklearn.svm import SVR
from sklearn.metrics import r2_score

linear_svr = SVR(kernel='linear', C=1.0)
linear_svr.fit(X_train, y_train)
y_pred = linear_svr.predict(X_test)
print('Model R2 score with linear kernel and C=1.0: {0:.4f}'.format(r2_score(y_test, y_pred)))
```

```python
from sklearn.svm import SVR
from sklearn.metrics import r2_score

linear_svr = SVR(kernel='linear', C=100.0)
linear_svr.fit(X_train, y_train)
y_pred = linear_svr.predict(X_test)
print('Model R2 score with linear kernel and C=100.0: {0:.4f}'.format(r2_score(y_test, y_pred)))
```

```python
from sklearn.svm import SVR
from sklearn.metrics import r2_score

poly_svr=SVR(kernel='poly',C=1.0)
poly_svr.fit(X_train,y_train)
y_pred=poly_svr.predict(X_test)
print('Model R2 score with polynomial kernel and C=1.0: {0:.4f}'.format(r2_score(y_test,y_pred)))
```
