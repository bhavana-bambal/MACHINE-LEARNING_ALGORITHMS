```python
before_rows = df.shape[0]
df = df.dropna()
after_rows = df.shape[0]

print("Before rows:", before_rows)
print("After rows:", after_rows)
print("The number of dropped rows are:", before_rows - after_rows)
```

```python
X = df.iloc[:, :-1]
print(X)
```

```python
from sklearn.model_selection import train_test_split
Y = df.iloc[:, -1]
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.3, random_state=42
)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
depths = [2, 4, 6, 8]
criteria = ["squared_error", "friedman_mse"]
```

```python
for criterion in criteria:
    for depth in depths:
        regressor = DecisionTreeRegressor(
            max_depth=depth,
            criterion=criterion,
            random_state=0
        )
        regressor.fit(X_train, Y_train)

        Y_pred = regressor.predict(X_test)

        print("Criterion:", criterion)
        print("Max Depth:", depth)
        print("Mean Squared Error:", mean_squared_error(Y_test, Y_pred))
        print("R2 Score:", r2_score(Y_test, Y_pred))
        print("----------------------------------")
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
best_regressor = DecisionTreeRegressor(max_depth=4, random_state=0)
best_regressor.fit(X_train, Y_train)

plt.figure(figsize=(10,10))
tree.plot_tree(best_regressor, filled=True)
plt.show()
```
